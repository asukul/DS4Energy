```python
# Analyze the dataset
print("Dataset Info:")
print(df_encoded.info())

print("\nSample Data:")
print(df_encoded.head())

# Check unique values in each column
print("\nUnique Values per Column:")
print(df_encoded.nunique())

# Drop columns with no variability (all FALSE)
df_encoded = df_encoded.loc[:, (df_encoded != 0).any(axis=0)]

# Save cleaned data for verification
df_encoded.to_csv("cleaned_data.csv", index=False)
print("Cleaned data saved as 'cleaned_data.csv'.")
```
